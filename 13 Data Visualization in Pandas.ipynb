### Data Visualization in Pandas

```python
#importing the libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Plotting graph with plot() method

```python
data=pd.Series(np.random.randn(1000).cumsum())
data
```

#### 1.  Line Plot 

```python
data.plot()
plt.show()
```

#### Another Example

```python
df1 = pd.DataFrame(np.random.randn(100, 4),columns=['a','b','c','d'])
df1 = df1.cumsum()
df1
```

```python
df1.plot()
plt.show()
```

#### 2. Bar Graph

```python
df2=pd.DataFrame(np.random.rand(7,4),columns=['a','b','c','d'])
df2
```

```python
df2.plot.bar()
plt.show()
```

#### 3. Histogram

```python
df = pd.read_csv('iris-data-clean.csv')
df.columns=["sepal_length","sepal_width", "petal_length", "petal_width", "species"]
```

```python
df.head()
```

```python
df.plot.hist(alpha=0.5)
plt.show()
```

#### 4. Boxplot Charts

```python
df.plot.box()
plt.show()
```

```python
# Changing the colors
colors={'boxes': 'Red', 'whiskers': 'blue','medians': 'Black', 'caps': 'Green'}
df.plot.box(color=colors)
plt.show()
```

#### To plot Horizontally

```python
df.plot.box(vert=False)
plt.show()
```

#### 5. Area Charts

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=['a','b','c','d'])
df.head()
```

```python
df.plot.area()
plt.show()
```

```python
#for plotting specified column.
df['c'].plot.area()
plt.show()
```

#### 6. Scatter Plots

```python
df.plot.scatter(x='a', y='b')
plt.show()
```

```python

df = pd.read_csv('iris-data-clean.csv')
df.columns=["sepal_length","sepal_width", "petal_length", "petal_width", "species"]
df.plot.scatter(x='sepal_length', y='sepal_width', s=df['petal_length'] * 50)
plt.show()
```

#### 7. Pie Charts


```python
df_avg=df["petal_width"].groupby(df["species"]).mean()
df_avg
```

```python
df_avg.plot.pie()
```

#### 8. Density Chart

```python
df.plot.kde()
plt.show()
```
